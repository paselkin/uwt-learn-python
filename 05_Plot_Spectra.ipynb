```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
%matplotlib inline

from specio import specread

# Find the path to the SP toy data
os.chdir('C:/Users/pasel/Dropbox/Research/IODP/Data/UV-Vis/2021-02-25_Scan/')
sp_glob = '*.sp'
for sp_filename in glob.glob(sp_glob):
    print(sp_filename)

    # Read the data
    spectrum = specread(sp_filename)

    # Plot the first spectrum
    fig, axs = plt.subplots(2, 1) 
    axs[0].plot(spectrum.wavelength, 100.*(1.-spectrum.amplitudes),'k-')
    axs[0].set_title(spectrum.meta['filename'])
    axs[0].set_ylabel('Reflectance (%)')
    axs[1].plot(spectrum.wavelength, np.gradient(100.*(1.-spectrum.amplitudes), spectrum.wavelength),'b-')
    axs[1].set_ylabel('First Derivative')
    axs[1].set_ylim(bottom=0.,top=1.)
    plt.savefig(spectrum.meta['filename'][:-3]+".pdf")
    plt.show()

    #input("Press Enter to continue...")
```
